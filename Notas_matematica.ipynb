## Notas **Aleatórias**

```python
from random import randrange
```

```python
notas_matematica = []
for notas in range(8):
  notas_matematica.append(randrange(0, 11))

notas_matematica
```

## **Gráfico**

```python
import matplotlib.pyplot as plt
```

```python
plt
```

```python
x = list(range(1, 9))
y = notas_matematica
plt.plot(x, y, marker='o')
plt.title('Notas de Matemática')
plt.xlabel('Provas')
plt.ylabel('Notas')
plt.show()
```

```python
def validacao():
  for notas in notas_matematica:
    if notas >=6 :
      print('Boa Nota, Parabéns')
    else:
      print('Péssima Nota, Melhore imediatamente')

validacao()
```
